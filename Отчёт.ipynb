# Отчёт по лабораторной работе № 5

## Введение

Лабораторна работа была направлена на проверку теоремы Штейнера. Формулировка теоремы:

$$ J = J_c + ma^2$$

Момент инерции тела относительно некоторой оси равен сумме собственного момента инерции тела относительно оси, проходящей черезх его центр масс и параллельной данной, и произведения массы на кавадрат расстояния между осями.

В ходе лабораторной работы измерялось время, за которое платформа совершает 30 колебаний. Использую эти данные считался период собственных колебаний платформы по формуле:

$$ T = \frac{t}{n}$$

Далее считался собственный момент инерции платформы по формуле:

$$J=\frac{MgRrT^2}{4\pi^2L}$$

M - масса платформы,
g - усорение свободного падения,
R - радиус нижней части платформы,
r - радиус верхней части платформы,
T - период колебаний платформы,
L - длина подвеса.

Далее проводились измерения времени, за которое совершает 30 колебаний система из платформу и двух цилиндров, расположенных симметрично на неотором расстоянии от центра масс платформы. Далее считался период колебаний и момент инерции системы по формулам выше.

Используя свойство аддитивности для каждого случая считался момент инерции одного цилиндра по формуле, выведенной с помощью свойства аддитивности моментов инерции:

$$ J_a = \frac{J' - J_0}{2}$$

J_a - момент инерции цилиндра относительно данной оси, J' - момент инерции ситемы, J_0 - моменте инерции платформы

Дальше был построен график зависимости момпента инерции цилиндра в зависимости от расстояния до центра масс. А также я посчитала относительную погрешность измерений.
